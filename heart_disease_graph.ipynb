# Task 1 - Find the diabetic and hight blood pressure patients across all age groups, and conclude the risk heart stroke is higher in which age group

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt 

#read the csv
df = pd.read_csv('heart_disease.csv')
df
```

```python
#Filter by diabetes(condition will be who has diabetes) and create new dataframe
diabetes = df.loc[df['diabetes'] == 1].reset_index()
diabetes
```

```python
#On this new data frame perform group operation as per age and create new dataframe 
age_group = diabetes.groupby('age')['diabetes'].count().reset_index()
age_group
```

```python
#Filter by high_blood_pressure(condition will be who has high_blood_pressure) and create new dataframe
high_blood_pressure = df.loc[df['high_blood_pressure'] == 1].reset_index()
high_blood_pressure
```

```python
#On this new data frame perform group operation as per age and create new dataframe 
age_group_bp = diabetes.groupby('age')['high_blood_pressure'].count().reset_index()
age_group_bp
```

```python
#plot the scatter graph to show which age group is more prone to diabetes
plt.figure(figsize=(19,8))
label = age_group['age']
value = age_group['diabetes']
plt.scatter(label,value,label="Number of People",color="maroon",alpha=0.5,s=value*100)

plt.xlabel("Age",fontsize=15)
plt.ylabel("No. of people prone to have diabetes",fontsize=15)
plt.title("NUMBER OF PEOPLE PRONE TO HAVE DIABETES IN VARIOUS AGES",fontsize=25)
plt.legend()

plt.show()
```

Conclusion - age group 60-70 has highest chance of having diabetes.

# Task 2 - Find as per gender who has not normal platelets level in blood

```python
#Filter by platelets(condition lesser then 150000 OR greater then 450000) and create new dataframe
abnormal_platelets = df.loc[(df['platelets'] < 150000) | (df['platelets'] > 450000)].reset_index()
abnormal_platelets
```

```python
#On this new dataframe perform group operation as per gender and create new dataframe 
gender = abnormal_platelets.groupby('gender')['platelets'].count().reset_index()
gender
```

```python
#Plot a pie chart as per the gender to show the percentage of male and female who has not normal platelets
gender['gender'] = ['female','male']
label = gender['gender']
value = gender['platelets']

plt.pie(value,labels=label,autopct="%0.2f%%",radius=2)
plt.show()
```

Conclusion - more males have abnormal platelets than females
